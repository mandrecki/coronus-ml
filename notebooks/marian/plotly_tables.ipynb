# Welcome plot
Active cases vs total cases?

```python
df_aggregations = pd.concat([
    df_active.sum(axis=1).rename("Active cases"),
    df_conf.sum(axis=1).rename("Total cases"),
    df_dead.sum(axis=1).rename("Deaths"),
    df_reco.sum(axis=1).rename("Recoveries")
], axis=1)
df_aggregations
```

```python
plot_interactive_df(df_aggregations[["Active cases", "Total cases"]], "number", " ")
```

# Digest

```python
total_cases_now = df_aggregations["Total cases"].iloc[-1]
cases_now = df_aggregations["Active cases"].iloc[-1]
cases_yesterday =  df_aggregations["Active cases"].iloc[-2]
new_cases = cases_now - cases_yesterday
percentage_growth = 100*new_cases/cases_yesterday
print("total cases now: {0:.0f}".format(total_cases_now))
print("active cases now: {0:.0f}".format(cases_now))
print("new cases: {0:.0f}".format(new_cases))
print("daily growth: {0:.1f}%".format(percentage_growth))
```
